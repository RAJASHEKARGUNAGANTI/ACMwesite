```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
data_set = pd.read_csv("/content/drive/MyDrive/DataSets/Crop_recommendation.csv")
```

```python
data_set
```

# New section

```python
data_set.isnull().any()
```

```python
x = data_set.drop("label",axis=1)
y = data_set["label"]
```

```python
x
```

```python

y
```

```python
x_train ,x_test ,y_train , y_test = train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=0)
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
Pridict = rf.predict(x_test)
```

```python
import joblib
```

```python
joblib.dump(rf, "crop_recommendation.sav")
```

```python
Pridict
```

```python
Accuracy= accuracy_score(y_test,Pridict)
```
